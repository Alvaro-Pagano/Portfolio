# 18 Regresion Lineal  con Statsmodels


```python
# Se utiliza como variables independientes a los ambientes contenidos en un inmueble
# para el caso de regresion lineal simple, para el caso de regresion lineal
# multiple se utilizara tambien la variable independiente Distancia que muestra la distancia
# promedio de cada casa a 5 distintos centros de empleo
```

```python
import statsmodels.api as sm
```

# 18.1.1 Definir variables dependientes y independientes

```python
# regresion lineal simple (x = Var.Independiente, y = Var.Dependiente)
x = laData ['Ambientes']
y = laData ['Valor (usd)']
```

# 18.1.2 Explorar el dataset

```python
laData.plot (kind = 'scatter',
            x='Ambientes',
            y='Valor (usd)')
```

# 18.1.3 Crear la regreion: Agrear constante y ajustar modelo

```python
x = sm.add_constant(x)
lm = sm.OLS(y,x).fit()
```

```python
lm.predict(x)
```

# 18.1.4 La Tabla de Regresion

```python
# tabla resumen de las metricas del modelo lineal 
lm.summary()
```

# 18.1.5 Ecuacion de la Regresion Lineal

```python
# Coef Ambientes: -1.003e+0

# Coef Constant: 5.404e+05

# Ecuacion lineal: y = ax + b

y_pred = -1.003e+0*x['Ambientes'] - 5.404e+05
```

# 18.1.6 Graficar la Regresion Lineal

```python
import seaborn as sns 
import matplotlib.pyplot as plt
# graficar los puntos 
plt.figure(figsize=(6,4), tight_layout=True)
sns.scatterplot(x=x['Ambientes'],y=y)
# graficar la linea
sns.lineplot(x=x['Ambientes'],y=y_pred, color='red')

# axis
plt.xlim(0)
plt.ylim(0)
plt.savefig('linear_regression')
plt.show()
```

# 19 Regresion Lineal Multiple

```python
# el booleano devuelve True cuando el valor que encuentra es 'Brasil'
# y False cuando el valor encontrado es distinto a Brasil 

df_Planilla["Nacionalidad"] == "Brasil"
```

```python
# forma del DataFrame
# (row, columns)

df_Planilla.shape
```

```python
# este booleano lo que hace es "recorrer" el DataFrame df_Planilla y en la columna especificada
# "Nacionalidad" analiza fila por fila para ver  si el valor encontrado es "Brasil", si esto pasa devuelve un booleano 
# True, si pasa lo contrario devuelve un False. Se define como una variable para utilizar estos
# resultados de True or False 


booleans = []
for length in df_Planilla["Nacionalidad"]:
    if length == "Brasil":
        booleans.append(True)
    else:
            booleans.append(False)
```

```python
booleans[0:15]
```

```python
df_Planilla[is_long]
```
